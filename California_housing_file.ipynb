# Data Loading

```python
from sklearn.datasets import fetch_california_housing
housing=fetch_california_housing()
X=housing.data
Y=housing.target
```

```python
X.shape
```

```python
Y.shape
```

```python
import pandas as pd
df=pd.DataFrame(X)
df
```

```python
df.columns=housing.feature_names
df
```

# DataPreprocessing

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# 2️⃣ Feature Engineering
df["Rooms_per_Household"] = df["AveRooms"] / df["AveOccup"]
df["Bedrooms_per_Room"] = df["AveBedrms"] / df["AveRooms"]
df["Population_per_Household"] = df["Population"] / df["AveOccup"]
df["Total_Rooms_Block"] = df["AveRooms"] * df["Population"]
```

```python
# 3️⃣ Remove Unnecessary or Highly Correlated Features
df.drop(["AveRooms", "AveBedrms", "AveOccup"], axis=1, inplace=True)
```

```python
# 4️⃣ Feature Scaling (Standardization)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df.iloc[:, :-1])  # Assuming last column is target
```

```python
scaled_df = pd.DataFrame(scaled_features, columns=df.columns[:-1])
```

```python
print(scaled_df.head())
```

```python
scaled_df
```

```python
X = scaled_df.iloc[:, :-1]  # Features
Y = scaled_df.iloc[:, -1]   # Target
```

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

# LinearRegression Model

```python
from sklearn.linear_model import LinearRegression
model_1=LinearRegression()
```

```python
model_1.fit(x_train,y_train)
```

```python
Y_pred=model_1.predict(x_test)
Y_pred
```

```python
import numpy as np 
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
r2 = r2_score(y_test, Y_pred)  # R² Score (Goodness of Fit)
mae = mean_absolute_error(y_test, Y_pred)  # Mean Absolute Error
rmse = np.sqrt(mean_squared_error(y_test, Y_pred))  # Root Mean Squared Error
print(f"R² Score: {r2:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
```

```python
# Create a DataFrame with predictions
predictions_df = pd.DataFrame({
    "Actual Values": y_test.values,  # Actual Prices
    "Predicted Values": Y_pred       # Predicted Prices
})
# Save to CSV file
predictions_df.to_csv("Linear_model_predictions.csv", index=False)
print("✅ Predictions saved to 'Linear_model_predictions'")
```

#  Ridge Regression (L2 Regularization) 

###### Ridge Regression is a type of linear regression that includes L2 regularization. This means that it adds a penalty term to the sum of squared coefficients, preventing the model from overfitting to the training data.It is useful when features are highly correlated (multicollinearity) because it shrinks coefficients, making the model more stable.



```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
model_2=Ridge()
param_grid = {'alpha': [0.001, 0.01, 0.1, 1, 10, 100]}  # Different values of alpha
grid_search = GridSearchCV(model_2, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(x_train, y_train)
```

```python
best_ridge = grid_search.best_estimator_
best_alpha = grid_search.best_params_['alpha']
best_ridge,best_alpha
```

```python
Y_pred_2=best_ridge.predict(x_test)
Y_pred_2
```

```python
import numpy as np 
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
r2 = r2_score(y_test, Y_pred_2)  # R² Score (Goodness of Fit)
mae = mean_absolute_error(y_test, Y_pred_2)  # Mean Absolute Error
rmse = np.sqrt(mean_squared_error(y_test, Y_pred_2))  # Root Mean Squared Error
print(f"R² Score: {r2:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
```

```python
# Create a DataFrame with predictions
predictions_df = pd.DataFrame({
    "Actual Values": y_test.values,  # Actual Prices
    "Predicted Values": Y_pred_2       # Predicted Prices
})
# Save to CSV file
predictions_df.to_csv("Ridge_model_predictions.csv", index=False)
print("✅ Predictions saved to 'Ridge_model_predictions'")
```
